```python
#Baixando dependências
import os
import pathlib
import requests
from requests_html import HTMLSession
from datetime import datetime
```

```python
#Preparando repositório para guardar arquivo bruto baixado pela Internet
path_raiz =  os.path.abspath('')
path_input =  path_raiz + '/Input/'
print(path_input)
```

```python
#Procurando Anuário Estatístico da ANP no ano mais recente
session = HTMLSession()

r = session.get("https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/anuario-estatistico-2023")
links = r.html.links
print(links)
```

```python
#Baixando tabela 1.1
nome_arquivo = "tabela1-1.csv"
for link in links:
    if str(link).find(nome_arquivo) != -1:
        r = requests.get(link, allow_redirects=True)
        with open(nome_arquivo, 'wb') as output:
            output.write(r.content)
        pathlib.Path(path_raiz + '/' + nome_arquivo).rename(path_input + nome_arquivo)
```
